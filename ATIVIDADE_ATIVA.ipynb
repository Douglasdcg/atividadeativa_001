```python
# Coffee Shops Tia Rosa - Sistema de Cafeteria
# Feito por: Douglas
# Descrição: Sistema de gerenciamento e pedidos


class Produto:
    """Classe de produto do estoque"""
    def __init__(self, nome, quantidade):
        self.nome = nome
        self.quantidade = quantidade

class Cliente:
    """Classe de cliente para fidelização"""
    def __init__(self, nome, email, telefone):
        self.nome = nome
        self.email = email
        self.telefone = telefone

# DADOS INICIAIS

# Estoque inicial {nome: objeto Produto}
estoque = {
    "Grãos de café": Produto("Grãos de café", 50),
    "Leite": Produto("Leite", 30),
    "Açúcar": Produto("Açúcar", 20),
    "Xaropes": Produto("Xaropes", 15),
    "Chocolate em pó": Produto("Chocolate em pó", 25)
}

# Lista de clientes
clientes = []

# Menu de bebidas
menu_bebidas = {
    1: "Café Americano",
    2: "Expresso",
    3: "Cappuccino",
    4: "Descafeinado",
    5: "Chocolate Quente"
}

# FUNÇÕES DE SISTEMA

def cadastrar_produto():
    """Cadastrar novo produto ou atualizar estoque"""
    nome = input("Nome do produto: ")
    qtd = int(input("Quantidade: "))
    if nome in estoque:
        estoque[nome].quantidade += qtd
    else:
        estoque[nome] = Produto(nome, qtd)
    print(f"Produto '{nome}' atualizado/cadastrado com sucesso!")

def listar_estoque():
    """Mostrar produtos no estoque"""
    print("\n Estoque Atual:")
    for produto in estoque.values():
        print(f"- {produto.nome}: {produto.quantidade} unidades")

def cadastrar_cliente():
    """Cadastrar cliente para fidelização"""
    nome = input("Nome: ")
    email = input("E-mail: ")
    telefone = input("Telefone: ")
    clientes.append(Cliente(nome, email, telefone))
    print(f"Cliente '{nome}' cadastrado com sucesso!")

def listar_clientes():
    """Listar clientes cadastrados"""
    print("\n Clientes Cadastrados:")
    for c in clientes:
        print(f"- {c.nome} | {c.email} | {c.telefone}")

def fazer_pedido():
    """Menu para o cliente escolher bebidas"""
    print("\n Bem-vindo ao Coffee Shops Tia Rosa!")
    print("=== MENU ===")
    for codigo, bebida in menu_bebidas.items():
        print(f"{codigo} - {bebida}")

    escolha = int(input("Escolha o número da bebida: "))
    if escolha in menu_bebidas:
        print(f"Você pediu: {menu_bebidas[escolha]}. Obrigado!")
    else:
        print("Opção inválida!")

# SISTEMAS

def sistema_funcionarios():
    """Menu para funcionários"""
    while True:
        print("\n=== SISTEMA DE GERENCIAMENTO ===")
        print("1 - Cadastrar produto")
        print("2 - Listar estoque")
        print("3 - Cadastrar cliente")
        print("4 - Listar clientes")
        print("0 - Voltar")
        op = input("Escolha: ")

        if op == "1":
            cadastrar_produto()
        elif op == "2":
            listar_estoque()
        elif op == "3":
            cadastrar_cliente()
        elif op == "4":
            listar_clientes()
        elif op == "0":
            break
        else:
            print("Opção inválida!")

def sistema_clientes():
    """Menu para clientes"""
    while True:
        print("\n=== SISTEMA DE PEDIDOS ===")
        print("1 - Fazer pedido")
        print("0 - Voltar")
        op = input("Escolha: ")

        if op == "1":
            fazer_pedido()
        elif op == "0":
            break
        else:
            print("Opção inválida!")

# PROGRAMA PRINCIPAL

def main():
    while True:
        print("\n=== Coffee Shops Tia Rosa ===")
        print("1 - Acesso Funcionários")
        print("2 - Acesso Clientes")
        print("0 - Sair")
        op = input("Escolha: ")

        if op == "1":
            sistema_funcionarios()
        elif op == "2":
            sistema_clientes()
        elif op == "0":
            print("Obrigado por usar o sistema!")
            break
        else:
            print("Opção inválida!")

if __name__ == "__main__":
    main()



```
